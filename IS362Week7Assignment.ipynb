```python
#import pandas and numpy per usual
import pandas as pd
import numpy as np
```

```python
#1. Load the ratings by user information that you collected into a pandas dataframe.
rated = pd.DataFrame(pd.read_csv('C://Users//danie//Documents//Movie Rated Information.csv'))
rated
```

```python
#2. Show the average ratings for each user and each movie.
rated_copy = rated.copy()
rated_copy['users average'] = rated_copy.mean(axis=1)
rated_copy
```

```python
movie_average = rated_copy[0:4].mean()
#adding columns for neater presentation
movie_average = rated_copy.append(movie_average, ignore_index=True)
movie_average.ix[5,'Users']="Movie Average"
movie_average
```

```python
#3. Create a new pandas dataframe, with normalized ratings for each user. 
#Again, show the average ratings for each user and each movie.

#according to http://bi-analytics.org/topic/9-standardization-vs-normalization/
#normalization = x(new) = x - x(min) / x(max) - x(min)
#standardization = x(new) = x - mean/ standard deviation
zscore1 = lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x))
zscore2 = lambda x: (x - x.mean()) / x.std()
```

```python
normalized = rated.copy()

cols_to_norm = ["Hidden Figures", "Beauty and the Beast (2017)", "John Wick 2", "Moana", "Jungle Book (2016)", "Doctor Strange"]
normalized = normalized[cols_to_norm].apply(zscore1)
normalized
```

```python
movie_average = normalized[0:4].mean()
movie_average = normalized.append(movie_average, ignore_index=True)
movie_average['users average'] = movie_average.mean(axis=1)
movie_average
#disadvanages: writing out the formula using lambda creates a series of nulls, unreliable information,
#very confusing and makes a mess
# If you have outliers in your data set, normalizing your data will certainly scale the “normal” data to a very small interval. 
#advantages: Normalization, which scales all numeric variables in the range [0,1].
#when done correctly, can be plotted on a graph and includes a large variety of information
#very important when dealing with parameters of different units and scales
```

```python
#[Extra credit] Create another new pandas dataframe, with standardized ratings for each user. 
#Once again, show the average ratings for each user and each movie.
standardized = normalized[cols_to_norm].apply(zscore2)
standardized
movie_average = standardized[0:4].mean()
movie_average = standardized.append(movie_average, ignore_index=True)
movie_average['users average'] = movie_average.mean(axis=1)
movie_average
```
